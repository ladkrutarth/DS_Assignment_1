```python

def print_structure(path, prefix=""):
    path = Path(path)
    print(prefix + "ðŸ“‚ " + path.name)
    for p in path.iterdir():
        if p.is_dir():
            print_structure(p, prefix + "    ")
        else:
            print(prefix + "    ðŸ“„ " + p.name)

print_structure("/content/Assignment_1/Farilty_study")
```

```python
plt.figure(figsize=(8,5))
# Select the 'math score' column for the histogram
sns.histplot(df['math score'], bins=20, kde=True, color="blue")
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
# Select the 'reading score' column for the histogram
sns.histplot(df['reading score'], bins=20, kde=True, color="blue")
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
# Select the 'Writing score' column for the histogram
sns.histplot(df['writing score'], bins=20, kde=True, color="blue")
plt.title("Distribution of Reading Scores")
plt.xlabel("Writing Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')
plt.show()
```
